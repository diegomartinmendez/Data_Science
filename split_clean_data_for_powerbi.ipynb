```python
data_amba_mesa = data_amba.drop(['distrito_id', 'distrito_nombre' , 'seccionprovincial_id' , 'seccion_id' , 'circuito_id' , 'mesa_tipo', 'mesa_electores', 'cargo_id', 'agrupacion_id' ]   , axis=1 , inplace=True)
```

```python
data_amba_mesa = data_amba[data_amba['votos_tipo'] == "POSITIVO" ]
data_amba_mesa = data_amba_mesa[data_amba_mesa['cargo_nombre'] == "PRESIDENTE Y VICE" ].fillna('CABA')
data_amba_mesa
```

```python
data_amba_mesa.drop(['cargo_nombre', 'agrupacion_nombre' , 'votos_tipo' ]   , axis=1 , inplace=True)
```

```python
data_amba_mesa
```

```python
total_mesa = data_amba_mesa.groupby(['seccionprovincial_nombre', 'seccion_nombre', 'circuito_nombre', 'mesa_id' ] ).aggregate("sum")
display(total_mesa)
```

```python
total_mesa.to_excel('total_mesa.xlsx')
```

```python
data_amba_circuito = data_amba_mesa
```

```python
data_amba_circuito.drop(['mesa_id'] , axis=1 , inplace=True)
```

```python
total_circuito = data_amba_circuito.groupby(['seccionprovincial_nombre', 'seccion_nombre', 'circuito_nombre'] ).aggregate("sum")
display(total_circuito)
```

```python
total_circuito.to_excel('total_circuito.xlsx')
```

```python
data_amba_partido = data_amba_circuito
```

```python
data_amba_partido.drop(['circuito_nombre'] , axis=1 , inplace=True)
```

```python
total_partido = data_amba_circuito.groupby(['seccionprovincial_nombre', 'seccion_nombre'] ).aggregate("sum")
display(total_partido)
```

```python
total_circuito.to_excel('total_partido.xlsx')
```
